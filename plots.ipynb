```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

csv = pd.read_csv("Resources/music.csv")
```

```python
csv
```

Year vs Duration

```python
yr_dur_df = csv[["year","duration"]]
yr_dur_df.head()
```

```python
yr_dur_df = yr_dur_df[yr_dur_df.year > 1960]
yr_dur_df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(yr_dur_df.year,yr_dur_df.duration,marker=".",alpha=.2)
plt.xlabel("Year")
plt.ylabel("Length of Song (seconds)")
plt.title("Length of Songs over Time \n since 1960")
z = np.polyfit(yr_dur_df.year,yr_dur_df.duration, 1)
p = np.poly1d(z)
plt.plot(yr_dur_df.year,p(yr_dur_df.year),"r")

plt.savefig("yr_duration.jpg")
plt.show()
```

Year vs Loudness

```python
yr_loud_df = csv[["year","loudness"]]
yr_loud_df.head()
```

```python
yr_loud_df = yr_loud_df[yr_loud_df.year > 1960]
yr_loud_df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(yr_loud_df.year,yr_loud_df.loudness,marker=".",alpha=.2)
plt.xlabel("Year")
plt.ylabel("Loudness of Song")
plt.title("Loudness of Songs over Time \n since 1960")
z = np.polyfit(yr_loud_df.year,yr_loud_df.loudness, 1)
p = np.poly1d(z)
plt.plot(yr_loud_df.year,p(yr_loud_df.year),"r--")
plt.savefig("yr_loudness.jpg")
plt.show()
```

North vs South

```python
no_so_df = csv[["latitude","tempo"]]
no_so_df.head()
```

```python
# remove 37.15735675 latitude point which is the default for Unknown location
no_so_df = no_so_df[no_so_df.latitude!= 37.15735675]
no_so_df.head()
```

```python
#continental US
no_so_df = no_so_df[no_so_df.latitude > 24]
no_so_df = no_so_df[no_so_df.latitude < 50]
no_so_df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(no_so_df.tempo,no_so_df.latitude,marker=".",alpha=.3)
plt.xlabel("Tempo (bpm)")
plt.axhline(40.71455,linewidth=1, color='k', linestyle="--")
plt.annotate('40.71455', xy=(235, 41))
plt.axhline(34.05349,linewidth=1, color='k', linestyle="--", label="34.05349")
plt.annotate('34.05349', xy=(235, 34.3))
plt.ylabel("Latitude")
plt.title("Tempo vs Latitude\n in the continental US")
plt.savefig("NorthSouth_tempo.jpg")
plt.show()
```

West vs East

```python
w_e_df = csv[["longitude","tempo"]]
w_e_df.head()
```

```python
# remove 37.15735675 latitude point which is the default for Unknown location
w_e_df = w_e_df[w_e_df.longitude!= -63.93335787]
w_e_df.head()
```

```python
#continental US
w_e_df = w_e_df[w_e_df.longitude > -124]
w_e_df = w_e_df[w_e_df.longitude < -62]
w_e_df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(w_e_df.longitude,w_e_df.tempo,marker=".",alpha=.3)
plt.axvline(-74.00712,linewidth=1, color='k', linestyle="--")
plt.annotate('-74.00712', xy=(-76,265),rotation=90, fontsize='small')
plt.axvline(-118.24532,linewidth=1, color='k', linestyle="--")
plt.annotate('-118.24532', xy=(-120,265),rotation=90, fontsize='small')
plt.xlabel("Longitude")
plt.ylabel("Tempo (bpm)")
plt.title("Tempo vs Longitude\n in the continental US")
plt.savefig("WestEast_tempo.jpg")
plt.show()
```
